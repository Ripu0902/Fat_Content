Changing % values to float values by just removing % as we can see those values also contain values more than 100% which gives hint that values might be something different than just being the constituent of the drink, so we are just going to remove % and then later we will scale that using min max scaler

```python
for col in train_pd.columns:
    if train_pd[col].dtype == 'object':
        if train_pd[col].str.contains('%').any():
            train_pd[col] = train_pd[col].astype(str).str.rstrip('%').astype('float')
```

Dropping columns such as Drink name, Drink Type, Flavour Variant, and Id as these column doesn't participate in calculting Fat_Content

```python
train_pd = train_pd.drop('Drink_Name', axis=1)
train_pd = train_pd.drop('Drink_Type', axis=1)
train_pd = train_pd.drop('Flavor_Variant', axis=1)
train_pd = train_pd.drop('Id', axis=1)
```

```python
train_pd
```

```python
train_pd.info()
```

Now as we can see abive some values are still string i.e, object so now let's map the Health Category using target encoding

```python
# Calculate the mean of the target variable for each category in 'Health_Category'
health_category_mapping = train_pd.groupby('Health_Category')['Fat_Content'].mean().to_dict()

# Apply target encoding
train_pd['Health_Category_Encoded'] = train_pd['Health_Category'].map(health_category_mapping)
```

```python
train_pd
```

Now dropping Health category Column

```python
train_pd = train_pd.drop('Health_Category', axis=1)
```

Also for the health rating as those values are ordinal we can just map those as 0 for low 1 for medium and 2 for high

```python
health_rating_mapping = {' Low ': 0, ' Medium ': 1, ' High ': 2}

train_pd['Health_Rating'] = train_pd['Health_Rating'].map(health_rating_mapping)
```

```python
train_pd
```

```python
train_pd.info()
```

Removing EMR rows in these Columns as below

```python
train_pd = train_pd[~train_pd['Sweetness_Index'].isin(['EMR'])]
train_pd = train_pd[~train_pd['Stimulant_Caffeine'].isin(['EMR'])]
train_pd = train_pd[~train_pd['Heart_Risk_Index'].isin(['EMR'])]
```

```python
columns_to_convert = train_pd.columns
```

Now changing the data types of all column to float

```python
for col in columns_to_convert:
    if col in train_pd.columns:
        try:
            train_pd[col] = train_pd[col].astype(float)
        except ValueError as e:
            print(f"Error converting column '{col}' to float: {e}")
```

```python
train_pd
```

```python
train_pd.info()
```

Resetting index so that we can devide data in to train and test split

```python
train_pd = train_pd.reset_index(drop=True)
```

```python
X = train_pd.drop('Fat_Content', axis=1)
y = train_pd['Fat_Content']
```

Now scaling DataFrame using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler


scaler_train_X = MinMaxScaler()


scaler_train_X.fit(X)


scaled_train_X = scaler_train_X.transform(X)


train_scaled_X = pd.DataFrame(scaled_train_X, columns=X.columns)
```

Splitting in Training and Testing in 4:1 ratio with random state 42

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_scaled_X, y, test_size=0.2, random_state=42) # 80% training and 20% test

# Now you have X_train, X_test, y_train, and y_test for your model
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Training an XG Boost Model and then calculating MSE and MAE

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error


model_xgb = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

# Train the model
model_xgb.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model_xgb.predict(X_test)

# Reshape predictions to match y_test
y_pred = y_pred.reshape(-1, 1)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```

Training Support Vector Regressor Mdoel  and calculating MSE and MAE

```python
from sklearn.svm import SVR


# Initialize the Support Vector Regressor model
svr_model = SVR(kernel='poly')

# Train the model
svr_model.fit(X_train, y_train)

# Make predictions on the test set
svr_pred = svr_model.predict(X_test)

# Evaluate the model
svr_mse = mean_squared_error(y_test, svr_pred)
print(f"SVR Mean Squared Error: {svr_mse}")
svr_mae = mean_absolute_error(y_test, svr_pred)
print(f"SVR Mean Absolute Error: {svr_mae}")
```

Now let's Train Artificial Neural Network for 50 epochs with 0.001 learning rate and then calculating MAE and MSE

```python
# prompt: apply ANN on this model with learning ratre

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the ANN model
model_ann = Sequential()
model_ann.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))  # Input layer
model_ann.add(Dense(32, activation='relu'))  # Hidden layer
model_ann.add(Dense(1))  # Output layer (regression)

# Compile the model with specified learning rate
learning_rate = 0.001  # You can adjust the learning rate
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
model_ann.compile(optimizer=optimizer, loss='mse', metrics=['mae'])


# Train the ANN model
model_ann.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2) # Adjust epochs and batch size

# Evaluate the ANN model
loss, mae = model_ann.evaluate(X_test, y_test)
print(f"ANN Mean Absolute Error: {mae}")
print(f"ANN Loss: {loss}")

# Make predictions using the trained model
ann_predictions = model_ann.predict(X_test)
```

Once trying PCA on the columns 'Saturated_Lipids', 'Cholesterol_Level' as PCA1 and 'Processed_Oil_Content', 'VitA_Percentage', 'Protein_Quantity', 'Carb_Count' as PCA2 because these hold high coorelation with each other

```python
from sklearn.decomposition import PCA

# Apply PCA
pca1 = PCA(n_components=1)
pca2 = PCA(n_components=1)

# Define features for PCA1
features_pca1 = ['Saturated_Lipids', 'Cholesterol_Level']
X_pca1 = X_train[features_pca1]
X_pca1_test = X_test[features_pca1]

# Apply PCA1
pca1_result = pca1.fit_transform(X_pca1)
pca1_result_test = pca1.transform(X_pca1_test)

# Define features for PCA2
features_pca2 = ['Processed_Oil_Content', 'VitA_Percentage', 'Protein_Quantity', 'Carb_Count']
X_pca2 = X_train[features_pca2]
X_pca2_test = X_test[features_pca2]

# Apply PCA2
pca2_result = pca2.fit_transform(X_pca2)
pca2_result_test = pca2.transform(X_pca2_test)


# Create new DataFrames with PCA components
X_train_pca = pd.DataFrame(X_train.drop(features_pca1 + features_pca2, axis=1))
X_train_pca['PCA1'] = pca1_result
X_train_pca['PCA2'] = pca2_result

X_test_pca = pd.DataFrame(X_test.drop(features_pca1 + features_pca2, axis=1))
X_test_pca['PCA1'] = pca1_result_test
X_test_pca['PCA2'] = pca2_result_test

# Train SVR model with PCA components
svr_model_pca = SVR(kernel='poly')
svr_model_pca.fit(X_train_pca, y_train)
svr_pred_pca = svr_model_pca.predict(X_test_pca)

# Evaluate the model
svr_mse_pca = mean_squared_error(y_test, svr_pred_pca)
print(f"SVR Mean Squared Error (with PCA): {svr_mse_pca}")
svr_mae_pca = mean_absolute_error(y_test, svr_pred_pca)
print(f"SVR Mean Absolute Error (with PCA): {svr_mae_pca}")
```
